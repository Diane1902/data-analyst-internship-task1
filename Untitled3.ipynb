```python
import pandas as pd
# Load the dataset
df = pd.read_csv('KaggleV2-May-2016.csv')

# First look at the data
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Check data types and missing values
print("Data Types & Missing Values:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

print("\nBasic statistics:")
df.describe()
```

```python
# Step 1: Rename columns to be clean and uniform
df.columns = ['patient_id', 'appointment_id', 'gender', 'scheduled_day',
              'appointment_day', 'age', 'neighbourhood', 'scholarship',
              'hypertension', 'diabetes', 'alcoholism', 'handicap',
              'sms_received', 'no_show']

print("✅ Columns renamed:")
print(df.columns.tolist())
```

```python
# Step 2: Fix date formats
df['scheduled_day'] = pd.to_datetime(df['scheduled_day']).dt.strftime('%d-%m-%Y')
df['appointment_day'] = pd.to_datetime(df['appointment_day']).dt.strftime('%d-%m-%Y')

print("✅ Date columns cleaned:")
print(df[['scheduled_day', 'appointment_day']].head())

# Step 3: Remove rows with negative age
print(f"\nRows with negative age: {(df['age'] < 0).sum()}")
df = df[df['age'] >= 0]
print(f"✅ Negative age rows removed!")
print(f"New dataset shape: {df.shape}")
```

```python
# Step 4: Standardize text columns
print("Gender values:", df['gender'].unique())
print("No_show values:", df['no_show'].unique())

# Make gender and no_show consistent (uppercase)
df['gender'] = df['gender'].str.strip().str.upper()
df['no_show'] = df['no_show'].str.strip().str.upper()

print("\n✅ After cleaning:")
print("Gender values:", df['gender'].unique())
print("No_show values:", df['no_show'].unique())
```

```python
# Step 5: Save the cleaned dataset
df.to_csv('medical_appointments_cleaned.csv', index=False)
print("✅ Cleaned dataset saved!")
print(f"Final shape: {df.shape}")

# Final summary
print("\n========= CLEANING SUMMARY =========")
print("1. Renamed all columns to lowercase with underscores")
print("2. Fixed date formats to dd-mm-yyyy")
print("3. Removed 1 row with negative age (-1)")
print("4. Standardized gender and no_show to uppercase")
print("5. No missing values found")
print("6. No duplicate rows found")
print("=====================================")
```
